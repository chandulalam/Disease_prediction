```python
# Distribution graphs (histogram/bar graph) of column data


def plotPerColumnDistribution(df1, nGraphShown=10, nGraphPerRow=3):
    nunique = df1.nunique()
    df1 = df1[[col for col in df1 if 1 < nunique[col] < 50]]  # Select columns with 1-50 unique values
    
    nRow, nCol = df1.shape
    columnNames = list(df1.columns)
    nGraphRow = max(1, (nCol + nGraphPerRow - 1) // nGraphPerRow)  # Ensure at least 1 row

    plt.figure(figsize=(6 * nGraphPerRow, 4 * nGraphRow), dpi=80)

    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df1.iloc[:, i]
        
        if pd.api.types.is_numeric_dtype(columnDf):
            columnDf.hist()
        else:
            columnDf.value_counts().plot(kind="bar")
        
        plt.ylabel('Counts')
        plt.xticks(rotation=90)
        plt.title(f'{columnNames[i]}')

    plt.tight_layout()
    plt.show()
```

```python
# Scatter and density plots


def plotScatterMatrix(df1, plotSize=10, textSize=8):
    df1 = df1.select_dtypes(include=[np.number])  # Keep only numerical columns
    df1 = df1.dropna(axis=1)  # Drop columns with NaN values
    df1 = df1[[col for col in df1 if df1[col].nunique() > 1]]  # Keep columns with more than 1 unique value

    columnNames = list(df1.columns)
    if len(columnNames) > 10:
        columnNames = columnNames[:10]  # Limit columns for better visualization

    df1 = df1[columnNames]
    
    ax = pd.plotting.scatter_matrix(df1, alpha=0.75, figsize=(plotSize, plotSize), diagonal='kde')

    corrs = df1.corr().values
    for i, j in zip(*np.triu_indices_from(corrs, k=1)):  # Use np.triu_indices_from correctly
        ax[i, j].annotate(f'Corr: {corrs[i, j]:.3f}', (0.8, 0.2), 
                          xycoords='axes fraction', ha='center', va='center', size=textSize)

    plt.suptitle('Scatter and Density Plot')
    plt.show()
```

```python
#list1 = DF['prognosis'].unique()
def scatterplt(disea):
    x = ((DF.loc[disea]).sum())#total sum of symptom reported for given disease
    x.drop(x[x==0].index,inplace=True)#droping symptoms with values 0
    print(x.values)
    y = x.keys()#storing nameof symptoms in y
    print(len(x))
    print(len(y))
    plt.title(disea)
    plt.scatter(y,x.values)
    plt.show()

def scatterinp(sym1,sym2,sym3,sym4,sym5):
    x = [sym1,sym2,sym3,sym4,sym5]#storing input symptoms in y
    y = [0,0,0,0,0]#creating and giving values to the input symptoms
    if(sym1!='Select Here'):
        y[0]=1
    if(sym2!='Select Here'):
        y[1]=1
    if(sym3!='Select Here'):
        y[2]=1
    if(sym4!='Select Here'):
        y[3]=1
    if(sym5!='Select Here'):
        y[4]=1
    print(x)
    print(y)
    plt.scatter(x,y)
    plt.show()
```
